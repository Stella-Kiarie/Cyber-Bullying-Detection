# Collecting Labeled Data

```python
import pandas as pd
import glob
import os
```

### Merging Labelled Dataset 

```python

folder_path = r'C:\Users\Lenovo\Desktop\Final Project Phase 5_Cyberbullying\Cyber-Bullying-Detection\Labelled_Data'

# Get the list of files
file_list = glob.glob(os.path.join(folder_path, "Dataset*.csv"))

if not file_list:
    print(f"Still no files found. Checking path: {folder_path}")
    if os.path.exists(folder_path):
        print("Folder exists, but no .csv files found inside.")
        print("Contents:", os.listdir(folder_path))
    else:
        print("Folder does not exist at this location.")
else:
    # Load and merge
    df_list = [pd.read_csv(file) for file in file_list]
    merged_df = pd.concat(df_list, ignore_index=True)
    print(f"Success! Merged {len(df_list)} files.")
    print(merged_df.info())
```

```python
merged_df.to_csv('Labelled_data.csv',index=False)
```
