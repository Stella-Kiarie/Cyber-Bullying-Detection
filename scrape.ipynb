```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

youtube = build('youtube', 'v3', developerKey="AIzaSyCtEXObNEafYof0XpO9dNlBPuz3mOkZYyc")

def get_comments_df(video_id):
    comments_list = []
    
    try:
        # Initial request
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100, # Max allowed is 100
            textFormat="plainText"
        )
        
        while request:
            response = request.execute()

            for item in response['items']:
                snippet = item['snippet']['topLevelComment']['snippet']
                comments_list.append({
                    'author': snippet['authorDisplayName'],
                    'text': snippet['textDisplay'],
                    'likes': snippet['likeCount'],
                    'published_at': snippet['publishedAt']
                })
            
            # This automatically handles the nextPageToken for you
            request = youtube.commentThreads().list_next(request, response)
            
            # Optional: Break if you reach your desired cap (e.g., 10,000)
            if len(comments_list) >= 10000:
                break

        return pd.DataFrame(comments_list)

    except HttpError as e:
        print(f"Error: {e}")
        return pd.DataFrame()
```
