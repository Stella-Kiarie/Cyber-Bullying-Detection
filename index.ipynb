```python
import pandas as pd
import numpy as np

# 1. Load your dataset
file_name = 'final_labeled_kenyan_dataset.csv'
dataset = pd.read_csv(file_name)

# 2. Split the dataset into 4 equal-ish parts
# array_split handles cases where the row count isn't exactly divisible by 4
portions = np.array_split(dataset, 4)

# 3. Save each portion to a new CSV file
for i, portion in enumerate(portions):
    output_name = f'member_{i+1}_dataset.csv'
    portion.to_csv(output_name, index=False)
    print(f"Saved {output_name} with {len(portion)} rows.")

print("\nSuccess! You now have 4 files ready for your group members.")
```
