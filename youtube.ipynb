```python
from googleapiclient.discovery import build
import pandas as pd

# Setup
API_KEY = "AIzaSyBmGXa7ZuJcx9rjVrSTUfKyShnH2vh3S04"
VIDEO_ID = "tzjA9bUZ9Yk"
youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_comments(v_id):
    all_comments = []
    # Initial request
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=v_id,
        maxResults=100,
        textFormat="plainText"
    )

    while request:
        response = request.execute()
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            all_comments.append({
                'author': comment['authorDisplayName'],
                'text': comment['textDisplay'],
                'likes': comment['likeCount'],
                'published_at': comment['publishedAt']
            })
        
        # Check if there's another page of comments
        request = youtube.commentThreads().list_next(request, response)

    return pd.DataFrame(all_comments)

# Run and Save
df = get_comments(VIDEO_ID)
df.to_csv("youtube_comments.csv", index=False)
print(f"Done! Saved {len(df)} comments to youtube_comments.csv")
```
